```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
numbers = np.load('numbers.npy')
without = np.load('without.npy')
sols = np.load('sols.npy')
```

```python
number = input()
```

```python
color = np.array([[0,0,180],[0,0,230],[0,200,0],[230,230,0],[250,250,50],[200,0,0],[250,0,0]])
for i in range(len(sols)):
    if numbers[i]==int(number):
        im = 0
        for j in range(len(sols[i])):
            im += sols[i][j][:,:, np.newaxis] * color[j].reshape(1,1,-1)
        plt.imshow(im)
        plt.show()
```
