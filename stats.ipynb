```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
numbers = np.load('numbers.npy')
without = np.load('without.npy')
sols = np.load('sols.npy')
```

```python
plt.hist(numbers,bins=31)
plt.show()
```

```python
plt.figsize= (40,40)
plt.plot(numbers, without, 'o')
plt.show()
```

```python
a = np.zeros((3,3))
a[:,0]=1
a[0,:]=1


b = np.zeros((4,2))
b[:,0]=1
b[0,:]=1

c = np.ones((2,3))
c[1,1] =0

d = np.ones((2,3))
d[1,2] =0

e= np.ones((2,4))
e[1,:] =0
e[1,1]=1

f= np.ones((2,4))
f[0,0] = 0
f[1,2] = 0
f[1,3] = 0

g= np.ones((3,3))
g[0,0] = 0
g[0,1] = 0
g[2,1] = 0
g[2,2] = 0

images = [a,b,c,d,e,f,g]
fig = plt.figure(figsize=(20, 10))

# Create a GridSpec with 2 columns, where the left column is divided into 6 rows
gs = fig.add_gridspec(7, 2, width_ratios=[1, 8], wspace=0)

for i in range(7):
    ax_left = fig.add_subplot(gs[i, 0])
    ax_left.imshow(images[6-i])  # Use imshow to display images
    ax_left.axis('off')  # Remove axis for cleaner image display

ax_right = fig.add_subplot(gs[:, 1])
ax_right.grid(True)
ax_right.set_xticks(np.arange(1,32))  # Ensure all x-ticks are visible
ax_right.set_yticks([])  # Remove y-ticks
ax_right.plot(numbers, without, 'x',  markersize=10, markeredgewidth=2, markeredgecolor='black')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
