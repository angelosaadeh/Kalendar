# Kalendar — exploration

Thin notebook over the `kalendar` package. The heavy lifting lives in `kalendar/`; here we just look at things.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
from matplotlib import pyplot as plt

from kalendar import solve, Solution
from kalendar.render import show_pieces, show_solution, show_stats
```

## The seven tiles

```python
show_pieces()
plt.show()
```

## Load precomputed solutions

Run `python scripts/solve.py` first to (re)generate `data/solutions.npz`.

```python
data = np.load('../data/solutions.npz')
grids, dates, unused = data['grid'], data['date'], data['unused_piece']
print(len(dates), 'solutions')
```

## Show a solution for a chosen day

```python
day = 14
idx = np.where(dates == day)[0][0]
sol = Solution(masks=None, grid=grids[idx], date=int(dates[idx]), unused=int(unused[idx]))
show_solution(sol)
plt.show()
```

## Stats: revealed date vs. unused tile

```python
show_stats(dates, unused)
plt.show()
```
